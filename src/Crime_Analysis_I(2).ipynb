Which year has the highest and lowest crime

```python
# year with the highest average crime
year_with_highest_crime = ave_crime_by_year.idxmax()
year_with_highest_crime
```

Which states have the highest and lowest violent crimes?

```python
state_crime = crime_dataset.groupby('agency_jurisdiction')['violent_crimes'].sum()
highest_state = state_crime.idxmax()
lowest_state = state_crime.idxmin()

print(f"State with highest violent crimes: {highest_state}")
print(f"State with lowest violent crimes: {lowest_state}")
```

```python
plt.figure(figsize=(12, 6))
state_crime.sort_values(ascending=False).head(10).plot(kind='bar', color='orange')
plt.title('Top 10 States by Violent Crimes')
plt.ylabel('Total Violent Crimes')
plt.xlabel('State')
plt.xticks(rotation=45)
plt.show()
```

```python
# ave crime in 1980
crime_1980 = crime_dataset[crime_dataset['report_year'] == 1980]

ave_crime_1980 =crime_1980['violent_crimes'].mean()
ave_crime_1980
```

Visualization

b. How have violent crimes changed over the years?

```python
# Group the data by 'report_year' and calculate the sum of 'robberies'
robberies_by_year = crime_dataset.groupby('report_year')['robberies'].sum()

# Plotting the data
plt.figure(figsize=(10, 6))
plt.plot(robberies_by_year.index, robberies_by_year.values, marker='o', linestyle='-', color='b', label='Robberies')

# Adding labels and title
plt.title('Robberies Throughout the Years', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Robberies', fontsize=12)
plt.grid(True)
plt.legend()
plt.tight_layout()

# Show the plot
```

Observation: declinig through the years

c. What is the distribution of robberies across states?

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='agency_jurisdiction', y='robberies', data=crime_dataset)
plt.title('Distribution of Robberies Across States')
plt.xticks(rotation=90)
plt.show()
```

bivariate analysis

Scatter Plot (e.g., violent crimes vs. robberies)

```python
crime_types_by_year = crime_dataset.groupby('report_year')[['robberies', 'violent_crimes']].sum()
crime_types_by_year.plot(kind='line', figsize=(12, 6), marker='o')
plt.title('Comparison of Crime Types Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Crimes')
plt.legend(title='Crime Types')
plt.grid(True)
plt.show()

```

 What is the trend of homicides per capita over the years?

```python
# Homicides per capita trend
homicides_trend = crime_dataset.groupby('report_year')['homicides_percapita'].mean()

# Plot
plt.figure(figsize=(10, 6))
plt.plot(homicides_trend.index, homicides_trend.values, marker='o', color='red', label='Homicides Per Capita')
plt.title("Homicides Per Capita Over the Years")
plt.xlabel("Year")
plt.ylabel("Average Homicides Per Capita")
plt.grid(True)
plt.legend()
plt.show()
```

 What is the distribution of crimes per capita across states?

```python
# Crimes per capita distribution
plt.figure(figsize=(12, 6))
sns.boxplot(x='agency_jurisdiction', y='crimes_percapita', data=crime_dataset)
plt.title("Distribution of Crimes Per Capita Across States")
plt.xlabel("State")
plt.ylabel("Crimes Per Capita")
plt.xticks(rotation=90)
plt.show()
```

 Is there a relationship between population and violent crimes?

```python
# Correlation
correlation = crime_dataset[['population', 'violent_crimes']].corr()
print("Correlation between Population and Violent Crimes:\n", correlation)

# Scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='population', y='violent_crimes', data=crime_dataset, alpha=0.7)
plt.title("Population vs Violent Crimes")
plt.xlabel("Population")
plt.ylabel("Violent Crimes")
plt.grid(True)
plt.show()
```

f. Which type of crime is most prevalent?

```python
# Total crimes by type
crime_types = crime_dataset[['homicides', 'rapes', 'assaults', 'robberies']].sum()

# Plot
crime_types.plot(kind='bar', figsize=(10, 6), color=['blue', 'green', 'orange', 'red'])
plt.title("Total Crimes by Type")
plt.xlabel("Crime Type")
plt.ylabel("Total Count")
plt.show()
```

g. How do crimes vary by population size categories?

```python
# Bin population into ranges
bins = [0, 100000, 500000, 1000000, 5000000, 10000000]
labels = ['<100K', '100K-500K', '500K-1M', '1M-5M', '>5M']
crime_dataset['population_bins'] = pd.cut(crime_dataset['population'], bins=bins, labels=labels)

# Total crimes by population bin
crimes_by_population = crime_dataset.groupby('population_bins')['violent_crimes'].sum()

# Plot
crimes_by_population.plot(kind='bar', color='purple', figsize=(10, 6))
plt.title("Violent Crimes by Population Size")
plt.xlabel("Population Size")
plt.ylabel("Total Violent Crimes")
plt.show()
```
